## Enrichment analyses
### Single

We first examined the univariate enrichment of SCZ risk variants in ASoC SNPs, OCRs and other annotated genomic regions, by using TORUS. Red dots and black bars represent enrichment odds ratio and 95% confidence interval, respectively. If the confidence interval does not include 0, we call the enrichment result statistically significant.

iN-Glut ASoC variants showed over 100-fold remarkable enrichment for SCZ risk variants, while NPC ASoC variants showed ~66-fold enrichment. In contrast, both iN-Glut and NPC OCRs revealed only ~3-fold enrichments (Fig 1). In other words, the function of neuronal ASoC iN-Glut and NPC variants are more highlighted than that in their corresponding OCRs. Other OCRs (iN-GA, iN-DN, iPSC) indicated significant enrichment as well. Nevertheless, the bar widths for ASoC variants are greater than those of OCRs, for the reason that the number of SCZ variants that overlap with ASoC SNPs is much fewer than those captured by OCR peaks.

It has been widely proven that SCZ risk loci are enriched in non-coding regulatory regions (1). We verified that conserved and 3'UTR sequences were slightly more enriched for SCZ risk variants than coding regions. Furthermore, SCZ risk variants were also highly enriched in 5'UTR and promoter sequences. We further performed single-annotation analysis using brain eQTL for comparison, since it has been demonstrated that aggregate associations between brain eQTLs and many brain disorder GWAS variants, such as SCZ (2). We discovered that brain eQTL revealed ~4-fold enrichment for SCZ risk variants (Fig 1). Specifically, it should be noted that repressed regions, also called inactive regions (3), showed negative enrichment (odds ratio less than 1) of SCZ risk variants, because it is strongly indicated that the genes within repressed regions are silenced (3).

We expanded similar enrichment analysis to 12 other complex diseases and traits (Fig. 2 A-L). As a result, we observed enrichment of GWAS variants for bipolar disorder (BP), educational attainment, height, intelligence, and neuroticism in ASoC SNPs — both iN-Glut and NPC (Fig. 2 EFGIJK). Enrichment in iN-Glut SNPs was also observed for Alzheimer’s disease (AD) and major depressive disorder (MDD) GWAS variants (Fig. 2 BJ). There were no enrichment of GWAS variants for attention deficit hyperactivity disorder (ADHD), autism spectrum disorder (ASD), body mass index (BMI), inflammatory bowel disease (IBD) and Parkinson's disease in neither iN-Glut or NPC ASoC SNPs (Fig. 2 ACDHL).

To improve visuality and comparability, we made bubble heatmap plots (Fig. 3 and 4) with size and color attributes for 13 GWAS traits and 15 genomic annotations in Figures 1 and 2. The color of the bubbles indicated the magnitude of enrichment (log2 odds ratio), while the size of the bubbles indicated the significance level. By default, the bubble with darker color and larger size denotes the enrichment result of larger magnitude and higher significance level.

In order to compare the enrichment of neuropsychiatric risk variants versus non-neuropsychiatric variants in neuronal ASoC SNPs and OCRs, we subsequently performed enrichment analysis for 9 control disorders or traits — head circumference, low-density lipoproteins (LDL), high-density lipoproteins (HDL), type 2 diabetes (T2D), total cholesterol, Crohn's disease, lupus, asthma and ulcerative colitis. As expected, control disorders or traits showed little or no enrichment in ASoC SNPs and OCRs (Fig. 5). For instance, the dots for LDL and T2D are light and small, indicating inconsiderable enrichment magnitude, or even slightly negative enrichment, as well as weak statistical significance. These results suggest that risk variants in neuropsychiatric disorders are likely to be linked to neuronal ASoC SNPs and OCRs, and highlight the advantage to utilize ASoC SNPs and OCRs in prioritizing non-coding region risk variant studies.

The red dots, representing neuropsychiatric disorders, concentrated at right side of both figures, especially top right corner (large odds ratio and high significance level). Meanwhile, most anthropometric, metabolic and immune diseases/traits located at the bottom part, which are statistically insignificant in spite of some occasional large odds ratios.
