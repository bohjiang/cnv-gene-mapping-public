```sos
get_pip_threshold <- function(rds_dat, p=0.05, interval=0.05, iter=1) {
    dat = readRDS(rds_dat)
    pip_threshold = c()
    while (TRUE) {
        pip_threshold[iter] = ifelse(p < 1, length(dat$pip[which(dat$pip < p)]) / length(dat$pip), break)
        iter = iter + 1
        p = p + interval
    }
    return (pip_threshold)
}
```

```sos
get_pip_threshold("data/del_sim_shape5_scale1_p0.01_N4000_M100_12.data.feather.RDS")
```
