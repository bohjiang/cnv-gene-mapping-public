## Background
Inflammatory bowel diseases (IBD) 
- IBD is a group of chronic gastrointestinal inflammatory disorders
- IBD affects millions worldwide, and is rising in prevalence
- IBD has two subtypes: ulcerative colitis (UC) and Crohn’s disease (CD)
- GWAS identified 200 IBD-associated loci, but few have been conclusively resolved to specific functional variants.
- linkage disequilibrium (LD) often results in similar evidence for association among many neighboring variants, difficult to resolve causal variants.

## Methods
### Fine-mapping method 1
Step 1: model selection of independent signals
- Bayesian information criterion (BIC)-based selection
- Penalized maximum likelihood across traits and variants
- Greedy search algorithm (steepest descent approximation)
- A swap-out repositioning adjustment

Step 2: construction of confident sets (CS)
- 95% CSs contain exactly one signal
- Conditional CS in the presence of multiple signals
- Potentially multiple CS generated per region

### Fine-mapping method 2
Step 1: use multinomial likelihood with correlated normal prior on the effect sizes
- Assume all loci are shared between traits
- Leverage the strong correlation in effect size
- directly calculates maximum apriori model estimates

Step 2: mapping secondary association signal
- Conditional regression followed by Step 1
- Stopping criteria: number of signals by method 1

### Fine-mapping method 3
Step 1: Bayesian LASSO
- Multiple causal variants jointly using a LASSO prior
- Sample from posterior rather than computing it
    - Sample y from liability threshold model
    - Sample configuration restricted to clusters
    - Sample parameters and hyperparameters (Gibbs)
- Get CS from posterior samples for each cluster

Step 2: cross trait CS construction
- Simply merge overlapping clusters

## Results

### Enrichment and colocalization
- Protein coding variants
    - 18-fold enrichment in non-synonymous
    - Only one variant with >50% certainty is synonymous
    - All previously reported IBD risk coding variants are included with >95% certainty

### Fine-mapping summary
- Among 97 IBD-associated loci, 94 identified consistent CS, 45 variants significantly enriched for protein-coding changes
- Size of CS: range from 1 to >400 variants
- 18 single causal variants with >95% certainty, of which 13 are novel
- 27 single causal variants with >50% certainty
- Most variants are associated with both UC and CD
- Number of potaneial candidate genes is reduced from 669 to 233

### Proportion of variance explained (PVE)
- CD: 25% CD risk explained by fine-mapping out of a total of 28% by all loci
- UC: 17% out of 22%

### Results of method 1
- Non-significant overlap of top PBMC eQTLs in CS identified
- Overlap of eQTLs for CD4+ cells and ileum cells observed
